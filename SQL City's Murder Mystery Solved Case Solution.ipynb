# Find out the Killer

```python
%%sql
SELECT *
FROM drivers_license
WHERE gender = 'male'
AND plate_number LIKE '%H42W%';
```

```python
%%sql
SELECT * 
FROM get_fit_now_member gm
JOIN get_fit_now_check_in gc
ON gm.id = gc.membership_id 
WHERE gm.membership_status = 'gold' 
AND gm.id LIKE '48Z%' 
AND gc.check_in_date = 20180109;
```

## Final Attempt to find Killer

```python
%%sql
SELECT *
FROM person p
JOIN get_fit_now_member gm ON p.id = gm.person_id
JOIN get_fit_now_check_in gc ON gc.membership_id = gm.person_id
JOIN drivers_license dl ON p.id = dl.id
WHERE gm.membership_status = 'gold' 
AND gm.id LIKE '48Z%' 
AND gc.check_in_date = 20180109
AND plate_number LIKE '%H42W%';
```

```python
%%sql
SELECT *
FROM person p
JOIN get_fit_now_member gm ON p.id = gm.person_id
JOIN get_fit_now_check_in gc ON gc.membership_id = gm.id
JOIN drivers_license dl ON p.license_id = dl.id
WHERE gm.membership_status = 'gold' 
AND gm.id LIKE '48Z%' 
AND gc.check_in_date = '20180109'
AND dl.plate_number LIKE '%H42W%'; 
```

# Find out the main Mastermind behind the murder

We have informations
- Gender : Woman
- Height : around 5'5" (65") or 5'7" (67")
- Hair : Red
- Car : Tesla Model S
- attended the SQL Symphony Concert 3 times in December 2017.

Relations we need -   
- person
- drivers_license
- facebook_event_checkin


```python
%%sql
SELECT *
FROM person p
JOIN drivers_license dl ON p.license_id = dl.id
JOIN facebook_event_checkin fe ON p.id = fe.person_id
WHERE dl.gender = "female"
  AND dl.height BETWEEN 65 AND 67
  AND dl.hair_color = "red"
  AND dl.car_model = "Model S"
  AND fe.event_name = "SQL Symphony Concert"
  AND fe.date LIKE '201712%'
GROUP BY p.name
HAVING COUNT(*) = 3;
```
